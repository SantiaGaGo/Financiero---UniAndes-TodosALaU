```python
#librerias y manejo de tablas
import yfinance as yfin
import pandas as pd
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt
import mplfinance as mpl
```

```python
#descargar dtos
df= yfin.download('TSLA', start='2020-01-01', end='2021-01-01')
df
```

```python
#descargar S&P500
df2= yfin.download('^GSPC', start='2020-01-01', end='2021-01-01')
df2
```

```python
#grafico
plt.figure(figsize=(16,8))
plt.title('empresa vs Mercado')
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre')
plt.plot(df['Close'], color='red')
plt.plot(df2['Close'], color='blue')
plt.legend(['Tesla', 'S&P500'], loc='lower right')
plt.show()
```

```python
#escalar la serie a 100 USD
df['Close_100']=(100/df['Close'][0])*df['Close']
df2['Close_100']=(100/df['Close'][0])*df2['Close']
df
```

```python
#grafico
plt.figure(figsize=(16,8))
plt.title('empresa vs Mercado')
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre escalados a 100 dolares')
plt.plot(df['Close_100'], color='red')
plt.plot(df2['Close_100'], color='blue')
plt.legend(['Tesla', 'S&P500'], loc='lower right')
plt.show()
```

```python
#Diagrama de velas japonesas
mpl.plot(df,
         type= 'candle',
         volume= True,
         figratio=(16,8),
         style='yahoo',
         title='velas japonesas empresa'
         )
```
