```python
import numpy as np
from scipy.stats import levene

# Generar datos simulados para tres grupos
np.random.seed(42)
grupo_A = np.random.normal(loc=50, scale=14, size=30)  # Media 50, Desviación estándar 14
grupo_B = np.random.normal(loc=52, scale=20, size=30)  # Media 52, Desviación estándar 20
grupo_C = np.random.normal(loc=55, scale=17, size=30)  # Media 55, Desviación estándar 17

# Prueba de Levene
stat, p_value = levene(grupo_A, grupo_B, grupo_C, center='median')

# Resultados
print(f"Estadístico de Levene: {stat:.4f}")
print(f"Valor p: {p_value:.4f}")

# Interpretación
alpha = 0.05
if p_value < alpha:
    print("Se rechaza la hipótesis nula: las varianzas no son similares.")
else:
    print("No se rechaza la hipótesis nula: las varianzas son homogéneas.")
```
